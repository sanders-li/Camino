```python
import json
from Sights_DB_DF import Sights_DB
import pandas as pd
import wikipedia
from nltk.tokenize import sent_tokenize 
import re
from bs4 import BeautifulSoup
from bs4.element import Comment
```

```python
with open('db_verification.json', 'r') as f:
    access_dict = json.load(f)
with Sights_DB(access_dict) as sights_db:
    df = sights_db.load_sights('Tokyo')
```

```python
paren_filter = re.compile(r'\s\([^)]*\)')
df.head()
```

```python
name = 'Imperial Palace'
try:
    wiki = wikipedia.WikipediaPage(name)
except wikipedia.DisambiguationError as e:
    wiki = wikipedia.page(e.options[0])
print(wiki.summary)
```

```python
summary = wiki.summary
x = ' '.join(sent_tokenize(summary)[:2])
summary = re.sub(paren_filter, '', x)
print(summary)
```
